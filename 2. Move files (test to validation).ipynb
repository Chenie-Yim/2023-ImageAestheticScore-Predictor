```python
with open("/Users/user/Desktop/Projects/CMSE890 - AML/image/fooddrink_vali_fin.txt", 'r') as f:
    ids = [line.rstrip() for line in f]
f.close()
```

```python
src = '/Users/user/Desktop/Projects/CMSE890 - AML/image/validation'
dest = '/Users/user/Desktop/Projects/CMSE890 - AML/image/validation'

for id in ids:
    image_file_name = id
    source_path = os.path.join(src, image_file_name +'jpg')
    if os.path.exists(source_path): 
        destination_path = os.path.join(dest, image_file_name +'.jpg')
        os.replace(source_path, destination_path)
        print(f'Moved {image_file_name} from test to validation.')
    else:
        print(f'{image_file_name} does not exist in the test folder.')
```
